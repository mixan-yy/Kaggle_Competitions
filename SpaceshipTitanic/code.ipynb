```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
df_train[:2]
```

```python
data = df_train.drop(['PassengerId','Name'], axis=1)
data_unseen = df_test.drop(['Name'], axis=1)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *
exp_mclf101 = setup(data = data, target = 'Transported', session_id=123, use_gpu = True) 
```

```python
best = compare_models()
```

```python
tuned_best = tune_model(best)
```

```python
predict_df = predict_model(tuned_best, data=data_unseen)
```

```python
submission = predict_df[['PassengerId', 'Label']]
```
